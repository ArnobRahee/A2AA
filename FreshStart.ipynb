```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score



# Convert 'date' column to datetime format
dfcombined['date'] = pd.to_datetime(dfcombined['date'])

# Select relevant columns for prediction
selected_columns = ['Industry', 'total_layoffs', 'date']
df = dfcombined[selected_columns]

# Filter data for the year 2024
df_2024 = df[df['date'].dt.year == 2023]

# Drop the original 'Industry' column before encoding
df_2024_encoded = pd.get_dummies(df_2024.drop('Industry', axis=1), columns=['date'])

# Split the data into features (X) and target (y)
X = df_2024_encoded
y = df_2024['Industry']

# Check if the dataset is sufficient for train-test split
if len(X) == 0 or len(y) == 0:
    print("Insufficient data for train-test split. Consider adjusting parameters or using alternative approaches.")
else:
    # Adjust train-test split ratio based on available data
    test_size = min(0.2, len(X) / (len(X) + len(y)))
    train_size = 1.0 - test_size

    # Split the data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, train_size=train_size, random_state=42)

    # Train a Random Forest classifier
    clf = RandomForestClassifier()
    clf.fit(X_train, y_train)

    # Predict the industry for the test set
    y_pred = clf.predict(X_test)

    # Evaluate the accuracy of the model
    accuracy = accuracy_score(y_test, y_pred)
    print("Accuracy:", accuracy)

    # Predict the industry with the highest layoffs in 2024
    predicted_industry = clf.predict(X)
    industry_with_highest_layoffs = pd.Series(predicted_industry).value_counts().idxmax()

    print("Industry with the highest layoffs probability in 2023:", industry_with_highest_layoffs)
```

```python
import pandas as pd
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt



# Convert 'date' column to datetime format
dfcombined['date'] = pd.to_datetime(dfcombined['date'])

# Select relevant columns for prediction
selected_columns = ['date', 'Industry', 'total_layoffs']
df = dfcombined[selected_columns]

# Group by 'date' and 'Industry' and calculate the sum of 'total_layoffs'
df_industry = df.groupby(['date', 'Industry']).sum().reset_index()

# Pivot the data to create a multi-variate time series
df_pivot = df_industry.pivot_table(index='date', columns='Industry', values='total_layoffs', aggfunc='sum')

# Resample the data to monthly frequency
df_pivot = df_pivot.resample('M').sum()

# Split the data into train and test sets
train_data = df_pivot[df_pivot.index.year < 2024]
# Adjust the test period to the available data
test_data = df_pivot[(df_pivot.index.year >= 2020) & (df_pivot.index.year < 2025)]

# Initialize a dictionary to store the SARIMAX models and predictions for each industry
models = {}
predictions = {}

# Fit SARIMAX models and make predictions for each industry
for industry in df_pivot.columns:
    # Fit the SARIMAX model
    order = (1, 1, 1)  # (p, d, q)
    seasonal_order = (1, 1, 1, 12)  # (P, D, Q, S)
    model = SARIMAX(train_data[industry], order=order, seasonal_order=seasonal_order, enforce_stationarity=False, enforce_invertibility=False)
    model_fit = model.fit()
    models[industry] = model_fit

    # Make predictions for the test set
    predictions[industry] = model_fit.predict(start=test_data.index[0], end=test_data.index[-1])

# Calculate root mean squared error (RMSE) for each industry
rmse_scores = {}
for industry in df_pivot.columns:
    industry_test_data = test_data[industry]
    industry_predictions = predictions[industry][industry_test_data.index]
    mse = mean_squared_error(industry_test_data, industry_predictions)
    rmse = mse ** 0.5
    rmse_scores[industry] = rmse

# Sort industries by RMSE scores in ascending order
sorted_industries = sorted(rmse_scores, key=rmse_scores.get)

# Print RMSE scores for each industry
for industry in sorted_industries:
    print(f"RMSE for {industry}: {rmse_scores[industry]}")

# Visualize the actual and predicted values for a specific industry
desired_industry = 'Energy'
plt.figure(figsize=(10, 6))
plt.plot(df_pivot.index, df_pivot[desired_industry], label='Actual')
plt.plot(predictions[desired_industry].index, predictions[desired_industry], label='Predicted')
plt.title(f"Layoffs in {desired_industry}")
plt.xlabel("Year")
plt.ylabel("Total Layoffs")
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm

# Assuming your dataset is already loaded into 'dfcombined'
# Make sure 'Date' column is in datetime format
dfcombined['Date'] = pd.to_datetime(dfcombined['Date'])

# Prepare the dataset for time series forecasting
# Group data by 'Industry' to forecast layoffs for each industry
grouped = dfcombined.groupby('Industry')

# Dictionary to store forecasts for each industry
forecasts_dict = {}

for industry, industry_data in grouped:
    # Create a time series from the 'total_layoffs' data for the current industry
    ts = industry_data[['Date', 'total_layoffs']].set_index('Date')

    # Resample the time series to a frequency of 'Y' (yearly)
    ts_resampled = ts.resample('Y').sum()

    # Check for any missing years and fill them with zeros
    min_year = ts_resampled.index.min().year
    max_year = ts_resampled.index.max().year
    all_years = pd.date_range(start=f'{min_year}-01-01', end=f'{max_year}-01-01', freq='Y')
    missing_years = set(all_years).difference(ts_resampled.index)

    for year in missing_years:
        ts_resampled.loc[year] = 0

    ts_resampled.sort_index(inplace=True)

    # Perform time series forecasting using statsmodels SARIMA model
    # You can adjust the order and seasonal_order parameters based on your data characteristics
    order = (1, 1, 1)
    seasonal_order = (1, 1, 0, 12)

    model = sm.tsa.SARIMAX(ts_resampled, order=order, seasonal_order=seasonal_order)
    results = model.fit()

    # Forecast for the next 2 years (2024 and 2025)
    forecast = results.forecast(steps=2)

    # Store the forecast in the dictionary
    forecasts_dict[industry] = forecast

    # Plot the forecast for visualization
    plt.figure(figsize=(10, 5))
    plt.plot(ts_resampled, label='Historical Data')
    plt.plot(forecast, label='Forecast')
    plt.title(f"Forecast for {industry} Industry")
    plt.xlabel("Year")
    plt.ylabel("Layoff Count")
    plt.legend()
    plt.show()

# Find the industry with the highest layoff count in 2024 and 2025
highest_layoff_industry_2024 = max(forecasts_dict, key=lambda x: forecasts_dict[x].iloc[0])
highest_layoff_industry_2025 = max(forecasts_dict, key=lambda x: forecasts_dict[x].iloc[1])

print(f"Industry with the highest layoff count in 2024: {highest_layoff_industry_2024}")
print(f"Industry with the highest layoff count in 2025: {highest_layoff_industry_2025}")
```
